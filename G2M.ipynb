<h3>Question1 - Average profit per km?</h3>
# profit = price charged - cost of trip

```python
#Adding Profit column
masterdata['Profit'] = masterdata['Price Charged'] - masterdata['Cost of Trip']
masterdata['Profit/km'] = masterdata['Profit'] / masterdata['KM Travelled']
masterdata['Profit Margin %'] = (masterdata['Profit'] / masterdata['Price Charged']) * 100
masterdata.head()
```

```python
#convert'Date of Travel to date time'
masterdata['Date of Travel'] = pd.to_datetime(masterdata['Date of Travel'])
masterdata.dtypes
```

```python
masterdata['Year']= masterdata['Date of Travel'].dt.year
masterdata['Month']= masterdata['Date of Travel'].dt.month
masterdata['Day']= masterdata['Date of Travel'].dt.day
masterdata.head()
```

```python
profitbykm = masterdata.groupby(['Year','Company'])['Profit/km'].mean()
profitperkm = pd.DataFrame({'Profit per Km':profitbykm}).reset_index()
profitperkm
```

```python
sns.barplot(x ='Year', y = 'Profit per Km',hue = 'Company',data = profitperkm).set_title('Profit per Km')
```

<h3>Question2 - profit percentage per year</h3>

```python
profitpercent = masterdata.groupby(['Year','Company'])['Profit Margin %'].mean()
profitmargin = pd.DataFrame({'Profit Margin%':profitpercent}).reset_index()
profitmargin
```

```python
sns.barplot(x ='Year', y = 'Profit Margin%',
            hue = 'Company',data = profitmargin).set_title('Proftit Margin %')
```

<h3>Question3 Gender based profit analysis</h3>

```python
profitpergender = masterdata.groupby(['Gender','Year','Company'])['Profit'].mean()
profitgender = pd.DataFrame({'Profit per gender':profitpergender}).reset_index()
profitgender
```

```python
plt.figure(figsize=(10, 8))
sns.catplot(x="Year", 
            y="Profit per gender", 
            hue="Gender", 
            col='Company',
            #order=['Pink cab', 'Yellow cab'],
            data=profitgender,
            kind="bar")

plt.savefig("grouped_barplot_Seaborn_barplot_Python.png")
```

<h3>Question4 Income class profit analysis</h3>

```python
def newdf(row):
    if row['Income (USD/Month)'] > 15000:
        val = 'high'
    elif (row['Income (USD/Month)'] < 15000) and (row['Income (USD/Month)'] > 2500):
        val = 'middle'
    else:
        val = 'low'
    return val

#create new column 'Class' using the function above
masterdata['Class'] = masterdata.apply(newdf, axis=1)

#view DataFrame 
masterdata.head()
```

```python
masterdata.head()
```

```python
economy = masterdata.groupby(['Year','Company','Class'])['Profit'].mean()
economic = pd.DataFrame({'Avg. Profit':economy}).reset_index()
economic
```

```python
plt.figure(figsize=(10, 8))
sns.catplot(x="Company", 
            y="Avg. Profit", 
            hue="Class", 
            col='Year',
            data=economic,
            kind="bar")

plt.savefig("grouped_barplot_Seaborn_barplot_Python1.png")
```

<h3>Question5 Profit by Age</h3>

```python
def newdf1(row):
    if row['Age'] >= 65:
        value = 'retired'
    elif (row['Age'] < 65) and (row['Age'] >= 41):
        value = 'senior'
    elif (row['Age'] < 40) and (row['Age'] >= 26):
        value = 'adult'
    else:
        value = 'young adult'
    return value

#create new column 'Class' using the function above
masterdata['Age group'] = masterdata.apply(newdf1, axis=1)

#view DataFrame 
masterdata.head()
```

```python
profitage = masterdata.groupby(['Year','Company','Age','Age group'])['Profit'].mean()
ageprofit = pd.DataFrame({'Avg. Profit':profitage}).reset_index()
ageprofit
```

```python
plt.figure(figsize=(10, 8))
sns.catplot(x="Company", 
            y="Avg. Profit", 
            hue="Age group", 
            col='Year',
            data=ageprofit,
            kind="bar")

plt.savefig("grouped_barplot_Seaborn_barplot_Python2.png")
```

<h3>Question6 City Users per Company</h3>

```python
usercity = masterdata.groupby(['City','Company'])['Users'].count()
cityusers = pd.DataFrame({'Users no.':usercity}).reset_index()
cityusers
```

```python
sns.set()
cities = cityusers
cities = cities.pivot("City", "Company", "Users no.")
ax = sns.heatmap(cities)
plt.title("City by Company Data")
plt.show()
```
